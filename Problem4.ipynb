```python
import numpy as np
import matplotlib.pyplot as plt

random_numbers = np.loadtxt(f"C:\\Users\\LENOVO\\Desktop\\Computational Physics 2024 feb-jun\\Assignment_solutions\\Assignment4\\exponential_random_numbers.txt")

plt.hist(random_numbers, bins=50, density=True, alpha=0.6, color='g')

x = np.linspace(0, max(random_numbers), 1000)
lambda_param = 2.0  # Because mean is 0.5, so lambda = 2
exponential_pdf = lambda_param * np.exp(-lambda_param * x)
plt.plot(x, exponential_pdf, 'r--')

plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Density Histogram of Exponential Random Numbers vs. Exponential PDF')
plt.legend(['Exponential PDF', 'Exponential Random Numbers'])

plt.show()
```
