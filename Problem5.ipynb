```python
import numpy as np
import matplotlib.pyplot as plt

def box_muller(n):
    u1 = np.random.rand(n // 2)
    u2 = np.random.rand(n // 2)
    z0 = np.sqrt(-2 * np.log(u1)) * np.cos(2 * np.pi * u2)
    z1 = np.sqrt(-2 * np.log(u1)) * np.sin(2 * np.pi * u2)
    return np.concatenate((z0, z1))


n_numbers = 10000
random_numbers = box_muller(n_numbers)

plt.hist(random_numbers, bins=50, density=True, alpha=0.6, color='g')

x = np.linspace(-4, 4, 1000)
gaussian_pdf = (1 / np.sqrt(2 * np.pi)) * np.exp(-0.5 * x**2)
plt.plot(x, gaussian_pdf, 'r--')

plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Density Histogram of Gaussian Random Numbers vs. Gaussian PDF')
plt.legend(['Gaussian PDF', 'Box-Muller Random Numbers'])

plt.show()
```
