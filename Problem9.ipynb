```python
import numpy as np
import matplotlib.pyplot as plt

def metropolis_sampling(n_samples, proposal_std, start_x=5):
    samples = []
    x = start_x 
    
    for _ in range(n_samples):
        x_proposed = np.random.normal(x, proposal_std)

        if 3 < x_proposed < 7:
            alpha = 1 
        else:
            alpha = 0  
        u = np.random.uniform(0, 1)
        if u < alpha:
            x = x_proposed
        
        samples.append(x)
    
    return np.array(samples)

n_samples = 10000
proposal_std = 0.5

samples = metropolis_sampling(n_samples, proposal_std)

plt.figure(figsize=(10, 4))
plt.plot(samples, lw=0.5)
plt.title('Markov Chain')
plt.xlabel('Iteration')
plt.ylabel('Sample Value')
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 4))
plt.hist(samples, bins=50, density=True, alpha=0.6, color='g')

x = np.linspace(3, 7, 1000)
uniform_pdf = np.ones_like(x) / (7 - 3)
plt.plot(x, uniform_pdf, 'r--')

plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Density Histogram of Generated Samples vs. Uniform Distribution')
plt.legend(['Uniform Distribution', 'Generated Samples'])
plt.grid(True)
plt.show()

```
