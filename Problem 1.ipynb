```python
import numpy as np
import matplotlib.pyplot as plt
import time


a = 1664525
c = 1013904223
m = 2**32
seed = 42

def lcg(n, seed, a, c, m):
    numbers = []
    x = seed
    for _ in range(n):
        x = (a * x + c) % m
        numbers.append(x / m)  # Scale to [0, 1]
    return numbers

n = 10000
t1 = time.time()
random_numbers = lcg(n, seed, a, c, m)
t2 = time.time()
print("time taken = ",t2-t1)
plt.hist(random_numbers, bins=50, density=True, alpha=0.6, color='b')

x = np.linspace(0, 1, 100)
uniform_pdf = np.ones_like(x)
plt.plot(x, uniform_pdf, 'r-')

plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Density Histogram of LCG Random Numbers vs. Uniform PDF')
plt.legend(['Uniform PDF', 'LCG Random Numbers'])
plt.show()
```
